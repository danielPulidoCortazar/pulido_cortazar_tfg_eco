# Análisis econométrico
*Parte de estimación seguida por medio de: Forecasting: Principles and Practice 2 ed: 5.2, 5.3.
Parte de predicción: seguida de 5.6*


$log \ E_t=(1-\phi) log \ \theta +\gamma_1(1-\phi) log \ PIB_t+\gamma_2(1-\phi) log \ p_t+ \phi \ log \ E_{t-1} + u_t$



En primer lugar vamos a comprobar que se cumplen las hipótesis del modelo clásico de regresión lineal. 



1.   Linealidad en los parámetros: se cumple por la especificación del modelo
2.   Muestreo aleatorio: suponemos que se cumple

3. $E(u_t | logPIB_t, log p_t, logE_{t-1})$ $→$ tenemos que comprobarlo
4. $Var(u_t \ | \ logPIB_t, log p_t, logE_{t-1}) = \sigma^2 \ ∀t$ $→$ tenemos que comprobarlo
5. $Cov(u_i, u_j \ | \ logPIB_t, log p_t, logE_{t-1}) \ ∀i,j \ i \neq j$ $→$ tenemos que comprobarlo
6. Ninguna de las variables es constante ni existe relación lineal exacta entre ellas. Se cumple, pues las variables no son transformaciones lineales entre si y son de naturaleza distinta.
7. El número de datos disponibles es mayor que el número de parámetros a estimar. Se cumple

Corremos la regresión con la función tslm()

```r
log.energia <- ts(variables[2:nrow(variables),"log.energia"], start = 1976)
log.pib <- ts(variables[2:nrow(variables),"log.PIB.real"], start = 1976)
log.precio.energia <- ts(variables[2:nrow(variables),"log.precio.energia"], start = 1976)
log.energia.lag1 <- ts(variables[2:nrow(variables),"log.energia.lag1"], start = 1976)
```

```r
ajuste.energia.especif1 <- tslm(
  log.energia ~ log.pib + log.precio.energia + log.energia.lag1)
summary(ajuste.energia.especif1)
```

log.PIB.real sale como poco significativo, probablemente debido a que la mayoría de la variabilidad positiva se la está llevando el consumo de energía corremos una regresión simple para comprobar si son significativos individualmente

El log.PIB.real resulta ser significativo en la regresión simple.

Visualizamos el ajuste de las 2 especificaciones

Analizamos los residuos para ver si se cumplen las hipótesis del modelo clásico

```r
cbind(Datos = variables[2:nrow(variables),"log.energia"],
      Ajuste = fitted(ajuste.energia.especif1)) %>%
  as.data.frame() %>%
  ggplot(aes(x=Datos, y=Ajuste)) +
    geom_point() +
    ylab("Valores ajustados") +
    xlab("Valores de la muestra") +
    ggtitle("Ajuste vs Muestra") +
    geom_abline(intercept=0, slope=1)
```

Veamos la función de autocorrelación de los residuos

```r
autoplot(Acf(residuals(ajuste.energia.especif1), plot = F))
```

Parece que la autocorrelación de los errores con sus valores pasados es pequeña. Terminemos de asegurar que se cumple la hipótesis de no autocorrelación
de los errores con el test de Breusch-Godfrey

```r
bgtest(ajuste.energia.especif1, order = 10)
```

```r
autoplot(residuals(ajuste.energia.especif1)) +
  theme(aspect.ratio = 1)
```

```r
checkresiduals(ajuste.energia.especif1)
```

Por el test Breush-Godfrey (BG) tenemos que fallamos en rechazar la hipótesis de que no hay autocorrelación hasta lag 9 incluido con un 10% de significación, luego parece razonable inferir que efectivamente cumplimos la hipótesis 5 de no autocorrelación de los errores.

A continuación vamos a comprobar que los erroes cumplen la hipótesis 4 de homocedasticidad. Una visualización de los residuos con respecto los valores ajustados no debería mostrar ningún tipo de patrón.

Efectivamente parece que no siguen ningún tipo de relación. Confirmemos nuestras sospechas formalmente con el contraste de Breusch-Pagan Test. 

```r
bptest(ajuste.energia.especif1)
```

Con un p-valor de 0.412 fallamos en rechazar la hipótesis nula de homocedasticidad con un 10% de significación. Luego tiene sentido inferir que se cumple la hipótesis 4 de homocedasticidad. Luego podemos confiar en los errores estándar de los parámetros. Luego se cumplen todas las hipótesis del modelo clásico de regresión lineal confirmando así que la regresión es consistente, insesgada y eficiente además de evitar la regresión espúrea de series temporales

# Modelo ARIMA PIB
*He seguido* *https://www.geo.fu-berlin.de/en/v/soga/Geodata-analysis/time-series-analysis/ARIMA-modelling-in-R/index.html*

```r
#generamos un objeto time serie tanto para los datos de entrenamiento como
#para los datos de prueba
pib <- ts(df["PIB.real.base2015.millones.euros"], start = 1975)
pib_entren <- ts(df["PIB.real.base2015.millones.euros"], start = 1975, end = 2017)
pib_test <- ts(df[43:48,"PIB.real.base2015.millones.euros"], start = 2018, end = 2021)
```

```r
print(pib_entren)
```

```r
print(pib_test)
```

```r
#representamos la serie
ggplot(df, mapping = aes(x = Periodo, y = PIB.real.base2015.millones.euros)) +
  ggtitle("PIB.real.base2015.millones.euros") +
  xlab("Año") +
  ylab("millones.euros") + 
  geom_line() +
  geom_point()
```

En el gráfico podemos observar que gracias a que está anualizada, la serie carece de estacionalidad pero parece seguir una tendencia incremental, luego es probable que no sea estacionaria y parece tener algunos ciclos también. Comprobamos la no estacionariedad con el test KPSS y el test Dickey-Fuller aumentado y el estadístico Ljung-Box (LB).

```r
#KPSS test
kpss.test(pib_entren)
```

Nos da un valor de 1.1487 luego podemos rechazar la hipótesis nula incluso al 1% de significación podemos rechazar la hipótesis nula de estacionaridad. Es decir, por medio de este test inferimos que nuestra serie es no estacionaria.

```r
#test Dickey-Fuller aumentado
adf.test(pib_entren)
```

Por otro lado con el test de Dickey Fuller aumentado podríamos rechazar la hipótesis nula de no estacionaridad con el 10% de significación. Según este test tendría sentido inferir que la serie es estacionaria alrededor de una tendencia determinista (gujarati). Necesitamos hacer un análisis un poco más exhaustivo.

```r
#estadístico Ljung-Box
Box.test(pib_entren)
```

El test realizado con el estadístico Ljung-Box nos permite inferir que la serie no se encuentra independientemente distribuida, esto es, que presenta correlación serial. Confirmémoslo definitivamente mostrando la función de autocorrelación

```r
#ACF
# funcion del paquete forecast con algunas mejoras con respecto a la función
# acf() de la base de R
autoplot(Acf(pib_entren, main = 'ACF for Differenced Series', plot = F))
```

La visualización del gráfico de autocorrelación nos confirma que se trata de una serie no estacionaria. Trataremos de hacerla estacionaria aplicando diferencias

```r
pib_entren_dif1 <- diff(pib_entren)
pib_entren_dif2 <- diff(pib_entren_dif1)
```

```r
autoplot(pib_entren_dif1) +
  geom_point()
```

Parece que hemos eliminado la tendencia. Volvemos a realizar los tests esta vez a la primera diferencia

```r
#kpss test
kpss.test(pib_entren_dif1)
```

El test KPSS toma un p-valor de 0.1, lo que nos hace fallar en rechazar la hipótesis nula al de estacionaridad

```r
adf.test(pib_entren_dif1)
```

Con el test Dickey-Fuller también fallamos en rechazar la hipótesis nula de estacionaridad.

```r
Box.test(pib_entren_dif1)
```

Por último y sin embargo, con el test del estadístico Ljung-Box rechazaríamos que los datos son independientemente distribuidos, siendo la hipótesis alternativa que no son independientes y que además representan correlación serial. Por lo tanto no podríamos rechazar la no estacionaridad. Es por ello que trataremos de ampliar el estudio por medio de la representación de la función de autocorrelación para comprobar si se extiende más allá esa supuesta correlación serial

```r
autoplot(acf(pib_entren_dif1,lag.max=40, plot = F))
```

```r
# funcion del paquete forecast con algunas mejoras con respecto a la función
# acf() de la base de R
autoplot(Acf(pib_entren_dif1, main = 'ACF for Differenced Series', plot = F))
```

En el gráfico vemos que existe una autocorrelación alta en el lag 1, y que decrece exponencialmente hasta convertirse en una onda senoidal de rango pequeño. Lo que nos puede hacer pensar que por esa razón el test LB no ha rechazado la hipótesis nula, pero que sin embargo podemos concluir que existe estacionaridad ya que en lags mayores que 1 la función de autocorrelación es similar a la del ruido blanco.

El siguiente paso es identificar el modelo correcto de la serie temporal. Hemos visto que es posible distinguir entre modelos AR, MA, ARMA por medio del comportamiento de los ACF y PACF.

Visualicemos ambos gráficos y veamos  cual es la configuración apropiada para el modelo

```r
#ACF
autoplot(Acf(pib_entren_dif1, plot = F, lag.max = 100)) + 
  ggtitle('ACF')
```

```r
#ACF
autoplot(Acf(pib_entren_dif1, plot = F, type = "partial", lag.max = 30)) + 
  ggtitle('PACF')
```

El ACF presenta una caida exponencial acompañada de variaciones de forma sinoidal a partir del lag 2. El PACF, por otro lado, presenta una caida abrupta después del lag 1 y tomando a partir de ese momento forma sinoidal. En base a los resultados del ACF y PACF parece lógico inferir ARMA(2,1) sobre la primera diferencia del pib, luego se trata elegimos el modelo ARIMA(2,1,1) para el pib (AR(2) para el pib_dif1 también puede tener sentido, según la figura 22.3 de gujarati)

Usamos la función Arima() que viene optimizada para el uso del paquete *forecast*

```r
pib_ajuste <- Arima(pib_entren, order = c(2,1,1), include.drift = T, seasonal = F)
summary(pib_ajuste)
```

sigma^2 se correspone con la varianza estimada del modelo. Luego el error estandar del modelo es:

```r
sqrt(224831679)
```

Ahora usamos la funcion auto.arima(), que usa una variacion de Hyndman and Khandakar algorithm que combina tests de raiz unitaria, minimización del AICc y estimación de máxima verosimilitud para obtener el modelo ARIMA más optimizado.

```r
pib_ajuste.automatico <- auto.arima(pib_entren, seasonal = F)
summary(pib_ajuste.automatico)
```

Vemos que la función auto.arima() estima un modelo arima óptimo acompañado de una tendencia *t*, tal y como el estadístico LB nos adelantaba cuando estudiabamos la estacionaridad. Además por medio del AIC podemos ver que se trata de un modelo más óptimo que el estimado manualmente.

Ahora nos queda hacer un análisis de los residuos y comprobar que efectivamente pueden ser considerados un proceso estocástico de ruido blanco, y por tanto no sería necesario buscar otro modelo ARIMA. Para ello aplicamos el test LB tests utilizados y visualizamos la función de autocorrelación y la de autocorrelación parcial para ambos ajustes.

```r
# LB test
Box.test(residuals(pib_ajuste))
Box.test(residuals(pib_ajuste.automatico))
```

En ambos casos nos da un p-valor no signficativo y por tanto fallamos en rechazar la hipótesis nula de errores distribuidos independientemente

```r
autoplot(Acf(residuals(pib_ajuste), plot = F))
autoplot(Acf(residuals(pib_ajuste.automatico), plot = F))
```

El correlograma nos confirma lo ya inferido en el test ya que la autocorrelación en diferentes lags presenta una forma similar a la que presentarían errores totalmente aleatorios (ruido blanco)

Procedemos a realizar la predicción con el modelo ajustado

```r
pib_ajuste_pred <- forecast(pib_ajuste, h = 20)
autoplot(pib_ajuste_pred, size = 1) + 
  autolayer(pib_test, size = 1, series = "muestra de prueba") +
    autolayer(fitted(pib_ajuste), size = 1, series = "ajuste" ) +
    theme(aspect.ratio = 1) +
    ylab("PIB real en millones de euros") +
    xlab("tiempo") +
    ggtitle("Predicción PIB real con ARIMA(2,1,1) con tendencia")

pib_ajuste.automatico_pred <- forecast(pib_ajuste.automatico, h = 20)
autoplot(pib_ajuste.automatico_pred) + 
  autolayer(pib_test)
```

```r
# muestra como se ajusta a los valores usando los datos de entrenamiento con un 
# determinado horizonte temporal "h"
autoplot(pib_entren, series="Training data") +
  autolayer(fitted(pib_ajuste.automatico, h = 2),
    series="N-step fitted values")
```

A continuación vamos a usar a computar el forecast usando los valores previos del test, es decir predecimos dentro de un horizonte temportal de un año.

```r
#predicción con horizonte temporal de un año sobre el test
pib_pred_1y <- Arima(pib_test, model=pib_ajuste.automatico)
accuracy(pib_pred_1y)
```

Pasamos a logaritmos los puntos predecidos tanto del precio como del pib y los convertimos en 2 series temporales

```r
log.pib_pred_st <- ts(log(pib_pred_df$Point.Forecast), start = 2022)
log.precio_energia_pred_st <- ts(log(precio_energia_pred_df$Point.Forecast[5:length(precio_energia_pred_df$Point.Forecast)]), #es toda la serie cuando usamos toda la muestra del precio de la energia
               start = 2022)
```

Integramos las predicciones con las series temporales de las muestras y generamos un data frame consolidad con todas las variables. 

```r
log.pib_st_consolidada <- ts(c(log.pib, log.pib_pred_st), start = 1975)
log.precio_energia_st_consolidada <- ts(c(log.precio.energia, log.precio_energia_pred_st), start = 1975)

#obtenemos el vector con máxima longitud, para así al crear el data frame, poder
# rellenar de NA los periodos que aún no han sido predecidos
longitud_max <-length(log.pib_st_consolidada)

length(log.pib_st_consolidada)
length(log.pib)
length(log.pib_pred_st)
length(log.precio_energia_st_consolidada)
length(log.precio.energia)
length(log.precio_energia_pred_st)

log.precio_energia_st_consolidada
```

```r
#generamos el data frame
variables_consolidada <- data.frame(log.energia = c(log.energia, rep(NA, longitud_max - length(log.energia))), 
                                  log.energia.lag1 = c(log.energia.lag1, rep(NA, longitud_max - length(log.energia.lag1))))
variables_consolidada$log.pib <- log.pib_st_consolidada
variables_consolidada$log.precio.energia <- log.precio_energia_st_consolidada
#generamos un vector de periodos para añadir al data frame
Periodos <- c(1976:2041)
variables_consolidada$Periodo <- Periodos
#cambiar el orden de las columnas para que los periodos estén al principio
variables_consolidada <- variables_consolidada[,c(5,1,2,3,4)]
#variables_consolidada <- variables_consolidada[ ,c(5,1,2,3,4)]
#generamos además una variable cualitativa que indique si es un valor estimado o real
variables_consolidada$estimado <- c(rep("dato",(length(1976:2021))), rep("estimación",length(2022:2041)))

variables_consolidada
```

Añadimos el valor de producción del año 2022 de energia a la variable log.energia.lag1 en la posición 2023 para poder comenzar a predecir

```r
variables_consolidada[variables_consolidada$Periodo == 2022, c("log.energia.lag1")] <- 7.348092

variables_consolidada
```

Generamos un bucle para calcular los nuevos valores de la energía que se utilizan para predecir los siguientes, hasta el horizonte 2042

```r
for (i in 2022:2040){
  periodo <- variables_consolidada$Periodo == as.character(i)
  variables_explicativas <- data.frame(log.pib = variables_consolidada[periodo,"log.pib"],
                                      log.precio.energia = variables_consolidada[periodo,"log.precio.energia"],
	                                    log.energia.lag1 = variables_consolidada[periodo,"log.energia.lag1"]
                                      )
  estimacion_i <- data.frame(forecast(ajuste.energia.especif1, newdata = variables_explicativas))
  variables_consolidada[periodo,"log.energia"] <- estimacion_i$Point.Forecast[1] 
  variables_consolidada[variables_consolidada$Periodo == as.character(i+1),"log.energia.lag1"] <- estimacion_i$Point.Forecast[1]
}
#para el año 2042, el ultimo año
periodo <- variables_consolidada$Periodo == as.character(2041)
variables_explicativas <- data.frame(log.pib = variables_consolidada[periodo,"log.pib"],
                                      log.precio.energia = variables_consolidada[periodo,"log.precio.energia"],
	                                    log.energia.lag1 = variables_consolidada[periodo,"log.energia.lag1"]
                                      )
estimacion_i <- data.frame(forecast(ajuste.energia.especif1, newdata = variables_explicativas))
variables_consolidada[periodo,"log.energia"] <- estimacion_i$Point.Forecast[1]
```
